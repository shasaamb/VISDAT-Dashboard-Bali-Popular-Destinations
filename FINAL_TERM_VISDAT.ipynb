## **Mapping Destinasi**

Convert Latitude, Longitude supaya pin point nya lebih presisi

```python
def dms_to_decimal(degree, minute, second, direction):
    decimal = degree + minute / 60 + second / 3600
    if direction in ['S', 'W']:
        decimal = -decimal
    return decimal
```

```python
def convert_to_decimal(coordinate):
    coordinate = coordinate.strip()
    if '°' in coordinate:  # Format: 8°37'16"S or 8.6212° S
        match_dms = re.match(r'(\d+)[° ](\d*)[′ ]?(\d*)[″ ]?([NSWE])', coordinate)
        match_decimal = re.match(r'(\d+\.\d+)[° ]([NSWE])', coordinate)

        if match_dms:
            degree, minute, second, direction = match_dms.groups()
            degree = int(degree)
            minute = int(minute) if minute else 0
            second = int(second) if second else 0
            return dms_to_decimal(degree, minute, second, direction)
        elif match_decimal:
            degree, direction = match_decimal.groups()
            decimal = float(degree)
            if direction in ['S', 'W']:
                decimal = -decimal
            return decimal
    return float(coordinate)
```

**berdasarkan rating**

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

bar_chart_data = data[['Place', 'Google Maps Rating', 'Google Reviews (Count)']].sort_values(by='Google Maps Rating', ascending=False)

def update_chart(min_rating, max_rating):
    filtered_data = bar_chart_data[(bar_chart_data['Google Maps Rating'] >= min_rating) &
                                   (bar_chart_data['Google Maps Rating'] <= max_rating)]
    fig_bar = px.bar(filtered_data, x='Place', y='Google Maps Rating',
                     title='Google Maps Ratings of Bali Tourist Destinations',
                     labels={'Place': 'Tourist Destination', 'Google Maps Rating': 'Rating'},
                     hover_data=['Google Maps Rating', 'Google Reviews (Count)'])
    fig_bar.show()

min_rating_slider = widgets.FloatSlider(value=bar_chart_data['Google Maps Rating'].min(),
                                        min=bar_chart_data['Google Maps Rating'].min(),
                                        max=bar_chart_data['Google Maps Rating'].max(),
                                        step=0.1,
                                        description='Min Rating')

max_rating_slider = widgets.FloatSlider(value=bar_chart_data['Google Maps Rating'].max(),
                                        min=bar_chart_data['Google Maps Rating'].min(),
                                        max=bar_chart_data['Google Maps Rating'].max(),
                                        step=0.1,
                                        description='Max Rating')

interactive_plot = widgets.interactive(update_chart, min_rating=min_rating_slider, max_rating=max_rating_slider)

display(interactive_plot)
```

```python
data['Region'] = data['Location'].apply(lambda x: x.split(',')[-1].strip())

pie_chart_data = data['Region'].value_counts().reset_index()
pie_chart_data.columns = ['Region', 'Count']

def update_pie_chart(selected_regions):
    filtered_data = pie_chart_data[pie_chart_data['Region'].isin(selected_regions)]
    fig_pie = px.pie(filtered_data, names='Region', values='Count',
                     title='Distribution of Tourist Destinations by Region in Bali',
                     labels={'Region': 'Region', 'Count': 'Number of Destinations'},
                     hover_data=['Region', 'Count'])
    fig_pie.show()

region_selector = widgets.SelectMultiple(
    options=pie_chart_data['Region'].unique(),
    value=list(pie_chart_data['Region'].unique()),
    description='Regions',
    disabled=False
)

interactive_pie = widgets.interactive(update_pie_chart, selected_regions=region_selector)

display(interactive_pie)
```
